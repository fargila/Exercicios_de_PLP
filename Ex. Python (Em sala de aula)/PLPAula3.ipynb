Programa 01 - Um progrma que controla uma caixa registradora. Solicite o código do produto e a qtd comprada. Utilize a seguinte tabela.

Código preço 1-> 0.50 | 2-> 1.00 | 3-> 4.00 | 5-> 7.00 | 9-> 8.00
Seu programa deve exibir o total das compras depois que o usuário digitar 0. Qualquer outro código deve gerar a mensagem "Código Inválido!".

```python
# Variáveis para o valor do código e a quantidade de produtos comprados >>
codigo = int(input("Informe o código do produto: "))
qtd = int(input("Informe a quantidade comprada: "))


# Variações dos preços de acordo com o código do produto num loop>> 
while(codigo != 0):

  if(codigo == 1):
    total = 0.50*qtd
    print("O valor total é: ")
    print(total)
  elif(codigo == 2):
    total = 1.00*qtd
    print("O valor total é: ")
    print(total)
  elif(codigo == 3):
    total = 4.00*qtd
    print("O valor total é: ")
    print(total)
  elif(codigo == 5):
    total = 7.00*qtd
    print("O valor total é: ")
    print(total)
  elif(codigo == 9):
    total = 8.00*qtd
    print("O valor total é: ")
    print(total)
  elif(qtd == 0):
    print("Quantidade nula!")
  else:
    print("Código Inválido!")

  codigo = int(input("Informe o código do produto: "))
  qtd = int(input("Informe a quantidade comprada: "))


```

```python
# Mesmo código acima, só que feito por meio de uma função >>
def soma(codigo, qtd):

  while(codigo != 0):

    if(codigo == 1):
      total = 0.50*qtd
      print("O valor total é: ")
      return print(total)
    elif(codigo == 2):
      total = 1.00*qtd
      print("O valor total é: ")
      return print(total)
    elif(codigo == 3):
      total = 4.00*qtd
      print("O valor total é: ")
      return print(total)
    elif(codigo == 5):
      total = 7.00*qtd
      print("O valor total é: ")
      return print(total)
    elif(codigo == 9):
      total = 8.00*qtd
      print("O valor total é: ")
      return print(total)
    elif(qtd == 0):
      return print("Quantidade nula!")
    else:
      return print("Código Inválido!")

codigo = int(input("Informe o código do produto: "))
qtd = int(input("Informe a quantidade comprada: "))

soma(codigo, qtd)
```
