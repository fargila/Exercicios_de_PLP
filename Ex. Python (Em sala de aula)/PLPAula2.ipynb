Programa 02 - Faça um programa que solicite o preço de um produto e o seu desconto, depois o preço a pagar

```python
def desconto(produto, desc):
  if(desc == 0):
    return print(produto)
  total = produto*((100-desc)/100)
  return print(total)

produto = float(input("Informe o preço do produto: "))
desc = float(input("Informe o desconto: "))

print("O desconto é: ")
desconto(produto, desc)
```

Programa 03 - Tempo de vida de um fumante. Pedir a qtd de cigarros fumados por dia e quantos anos já vem fumando, perde-se 10min de vida à menos a cada cigarro.



```python
def vida_fumante(qtdCig, anosFum):
  AnosFMin = (365*24*60)*anosFum
  temPerT = AnosFMin/(qtdCig*10)

  total = temPerT/24
  return total

qtd = float(input("Quantos cigarros você fuma por dia? "))
anos = float(input("Quantos anos vem fumando? "))

print("Total de dias perdidos: ")
vida_fumante(qtd, anos)


```
