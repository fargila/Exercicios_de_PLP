lê um vetor de 8 posições e depois, lê dois valores X e Y quaisquer correspondentes a duas posições no vetor. O programma deve imprimir a soma
dos valores encontrados nas respectivas posições X e Y.



```python

def main():
    vetor = [0] * 8

    for i in range(8):
        vetor[i] = int(input("Digite o valor da posição {}: " .format(i)))

    x = int(input("Digite o valor de X: "))
    y = int(input("Digite o valor de Y: "))
    if(0>x>7 or 0>y>7):
        print("O valor de X ou o valor de Y não pode ser maior que 8 nem menor que 0!")
    else:
        soma = vetor[x] + vetor[y]
        diminuicao = vetor[x] - vetor[y]
        print("A soma dos valores das posições X e Y é {}." .format(soma))
        print("A subtração dos valores das posições X e Y é {}." .format(diminuicao))

if __name__ == "__main__":
    main()
```

Faça um programa que receba do usuário
um vetor com 10 posições. Em seguida devera ser
impresso o maior e o elemento do vetor.
A soma de todos os elemnetos, a média e a qtd de elementos >=5

```python
def main():
    vetor = [0] * 10
    soma=0
    cont=0

    for i in range(10):
      vetor[i] = int(input("Digite o valor da posição {}: " .format(i)))
      soma += vetor[i]

    maior = vetor[0]
    menor = vetor[0]
    for i in range(10):
      if(vetor[i] > maior):
        maior = vetor[i]
      if(vetor[i] < menor):
        menor = vetor[i]
      if(vetor[i] >= 5):
        cont = cont + 1

    media = soma/10
    print("O maior número do vetor é {}." .format(maior))
    print("O menor número do vetor é {}." .format(menor))
    print("A soma total do vetor é {}." .format(soma))
    print("A média do vetor é {}." .format(media))

if __name__ == "__main__":
    main()
```
